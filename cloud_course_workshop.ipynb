## Introduction to Cloud Computing and Google Colab (30 minutes)

### 1. What is Cloud Computing and Google Colab?

- **Cloud Computing**:
  - *Definition*: Cloud computing is the delivery of computing services like storage, processing power, databases, and software over the internet, on a pay-as-you-go basis.
  - *How it works*: Instead of having physical servers and software installed on your local machine, you access these resources through the internet from a cloud service provider like Google, Amazon, or Microsoft.
  - *Examples*: Some popular cloud computing services include Google Cloud Platform, Amazon Web Services, and Microsoft Azure.

- **Google Colab**:
  - *Definition*: Google Colab (short for Colaboratory) is a free, cloud-based service provided by Google that allows you to write, execute, and share Python code in an interactive Jupyter Notebook-like environment.
  - *How it works*: You can create, edit, and share Colab notebooks using your Google account. Notebooks are stored on Google Drive and can be shared with others just like Google Docs or Sheets. Colab provides access to a virtual machine with pre-installed Python libraries and GPU/TPU support for running your code.

### 2. Benefits of using Cloud Computing and Google Colab

- **Cost-effective**: Save on resources with pay-as-you-go cloud services and enjoy free access to Google Colab.
- **Scalability and Flexibility**: Effortlessly scale resources in the cloud and switch between CPU, GPU, and TPU in Google Colab.
- **Collaboration**: Collaborate in real-time using cloud-based services and share notebooks in Google Colab.
- **Accessibility**: Access cloud resources and Google Colab from any device with an internet connection.


### 3. Use cases of Google Colab

- **Education and Learning**: Create and share interactive tutorials, and learn Python, data science, and more.
- **Data Analysis and Visualization**: Import, analyze, and visualize data using popular Python libraries.
- **Machine Learning and Deep Learning**: Build, train, and test models using popular frameworks and Colab's GPU/TPU support.
- **Prototyping and Experimentation**: Test ideas and collaborate on projects without setting up a local development environment.

```python
### import python libraries
import pandas as pd
import requests
import os
```

```python
# load the data we have saved from the storage account
raw_path = "/content/drive/MyDrive/ICTC/cloud_course/data/raw/"
data_1 = pd.read_csv(raw_path+"data1_2020_2022.csv")
data_2 = pd.read_excel(raw_path+"data2_2023.xlsx")
```

### Task to do


*   Extract only Province, WeeklyCases, WeeklyDeaths, WeekEnding
*   Create a folder structure on storage account in following way and save the data in csv format
image.png







```python
## merge the two datasets
covid_data = pd.concat([data_1, data_2])
```

```python
## filter the dataframe to select required columns
covid_data = covid_data[["Province", "WeeklyCases", "WeeklyDeaths", "WeekEnding"]]
```

```python
# preview the data
covid_data.head(2)
```

```python
# convert the WeekEnding column to datetime format
covid_data['WeekEnding'] = pd.to_datetime(covid_data['WeekEnding'])
```

```python
# extract the year from the WeekEnding column
covid_data['Year'] = covid_data['WeekEnding'].dt.year
```

```python
# preview the data
covid_data.head(2)
```

```python
# confirm we got all the years
covid_data.Year.unique()
```

```python
# check the unique provinces
covid_data.Province.unique()
```

```python
# make a list of unique years and provinces to create folders in storage account
year_list = list(covid_data.Year.unique())
province_list = list(covid_data.Province.unique())
```

```python
# define the path to the processed folder where the data to be saved
processed_folder = '/content/drive/MyDrive/ICTC/cloud_course/data/processed/'
for year in year_list:
  # create a folder with the year name inside the parent folder
  year_folder = os.path.join(processed_folder, str(year))
  os.makedirs(year_folder, exist_ok=True)
  # loop through each country in the countries list
  for province in province_list:
    # create a folder with the country name inside the year folder
    country_folder = os.path.join(year_folder, province)
    os.makedirs(country_folder, exist_ok=True)
    # filter the dataset to current province and year
    covid_data_filtered = covid_data[(covid_data.Year==year) & (covid_data.Province==province)]
    # save the filtered data back to the storage account
    file_name = province+"_"+str(year)+".csv"
    print(f"saving processed file at {country_folder} : {file_name}")
    covid_data_filtered.to_csv(country_folder+"/"+file_name)
```

#### **Festival Data API**
**about the data** : This dataset lists all festivals, special events and exhibits (etc.) approved to appear on the City of Toronto Festivals and Events Calendar.

All details are submitted by the event's organizer/organization, reviewed by City of Toronto - Tourism Services staff (against established submission criteria) and approved for posting to the calendar. The City is not responsible for the accuracy of the self-identified data provided by the organizations.<br>
URL : https://open.toronto.ca/dataset/festivals-events/
### Things to do.


*   Retrive JSON API url from the metadata
*   Parse the JSON data and extract required informations
* We will be extracting following informations
  * event name
  * category name
  * contact number
  * event date
  * event description
  * event location
  * features
  * organization name
  * organization type
  * reservation required
  * event website
* Create a dataframe using pandas
* Write the dataframe to storage account in excel format



```python
import requests
# Toronto Open Data is stored in a CKAN instance. It's APIs are documented here:
# https://docs.ckan.org/en/latest/api/

# To hit our API, you'll be making requests to:
base_url = "https://ckan0.cf.opendata.inter.prod-toronto.ca"

# Datasets are called "packages". Each package can contain many "resources"
# To retrieve the metadata for this package and its resources, use the package name in this page's URL:
url = base_url + "/api/3/action/package_show"
params = { "id": "festivals-events"}
package = requests.get(url, params = params).json()
festival_api = ""
# To get resource data:
for idx, resource in enumerate(package["result"]["resources"]):

       # To get metadata for non datastore_active resources:
       if not resource["datastore_active"]:
           url = base_url + "/api/3/action/resource_show?id=" + resource["id"]
           resource_metadata = requests.get(url).json()
           if resource_metadata['result']['format'] == 'JSON':
           # get the JSON api url
            festival_api = resource_metadata['result']['url']
```

```python
festival_api
```

```python
# make http call and convert the result to json format
response = requests.get(festival_api).json()
```

```python
len(response)
```

```python
# look at sample data
response[0]
```

```python
# create empty lists to store extracted data
category_list = []
date_list = []
description_list = []
event_name_list = []
location_list = []
org_name_list = []
phone_list = []
features_list = []
orgType_list = []
reservation_list = []
website_list = []
```

```python
# iterate through each event in the 'res' list
for event in response:
    # extract and append categoryString from calEvent to category_list
    category_list.append(event['calEvent']['categoryString'])

    # extract and append dates from calEvent to date_list
    date_list.append(event['calEvent']['dates'])

    # extract and append description from calEvent to description_list
    description_list.append(event['calEvent']['description'])

    # extract and append eventName from calEvent to event_name_list
    event_name_list.append(event['calEvent']['eventName'])

    # extract and append locationName from locations to location_list
    location_list.append(event['calEvent']['locations'][0]['locationName'])

    # extract and append orgName from calEvent to org_name_list
    org_name_list.append(event['calEvent']['orgName'])

    # try to extract and append eventPhone from calEvent to phone_list, if not available append 'N/A'
    try:
        phone_list.append(event['calEvent']['eventPhone'])
    except:
        phone_list.append("N/A")

    # try to extract and append features from calEvent to features_list, if not available append 'N/A'
    try:
        features_list.append(event['calEvent']['features'])
    except:
        features_list.append("N/A")

    # try to extract and append orgType from calEvent to orgType_list, if not available append 'N/A'
    try:
        orgType_list.append(event['calEvent']['orgType'])
    except:
        orgType_list.append("N/A")

    # try to extract and append reservationsRequired from calEvent to reservation_list, if not available append 'N/A'
    try:
        reservation_list.append(event['calEvent']['reservationsRequired'])
    except:
        reservation_list.append("N/A")

    # try to extract and append eventWebsite from calEvent to website_list, if not available append 'N/A'
    try:
        website_list.append(event['calEvent']['eventWebsite'])
    except:
        website_list.append("N/A")
```

```python
# create a dataframe using the list of extracted event details
eventDf = pd.DataFrame({
                        'EventName':event_name_list,
                        'Category':category_list,
                        'OrgName':org_name_list,
                        'OrgType':orgType_list,
                        'LocationName':location_list,
                        'Date':date_list,
                        'Contact':phone_list,
                        'Description':description_list,
                        'Features':features_list,
                        'ReservationRequired':reservation_list,
                        'Website':website_list
                      })
```
